```python
import gpt_2_simple as gpt2
from datetime import datetime
```

```python
import tensorflow as tf
import pandas as pd
import os
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import random
import re
```

```python
gpt2.download_gpt2(model_name="124M")
```

```python
# Datensatz einlesen
df = pd.read_csv('data/out.csv')
## Transform dataframe
# Dataframe mit 3 Spalten. Werden so gejoint, dass ein neues Dataframe mit ein Haiku pro Zeile erstellt wird
df = df[['0', '1', '2']].agg(lambda x: '\n'.join(x.values), axis=1)
# Dataframe to list [[]] -> []
array_of_poems = df.values.tolist()
# num_poems an Haikus mit Semicolon zwshceneinander zusammenfuegen
num_poems = 100000

text = ';\n\n'.join(array_of_poems[:num_poems])
```

```python
text_file = open("sample.txt", "w")
n = text_file.write(text)
text_file.close()
```

```python


sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset="sample.txt",
              model_name='124M',
              steps=1000,
              restore_from='fresh',
              run_name='run1',
              print_every=50,
              sample_every=50,
              save_every=500,
              )
```

```python
#sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, run_name='run1')
```

```python
gpt2.generate(sess,
              length=250,
              temperature=0.7,
              prefix="my",
              nsamples=5,
              batch_size=5
              )
```
